```python
from typing import Container
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'
headers = {'User-Agent':user_agent,} 

# url = ""
url = "https://elcinema.com/work/1014880/"

request = urllib.request.Request(url,None,headers)
response = urllib.request.urlopen(request)

html = response.read()
soup = bs(html, "html.parser")
# print(soup)
```

# Title

```python
post_title = soup.find_all("div",{"class": "jumbo"})
post_titles = post_title[0].findAll("h1")
post_titless = post_titles[0].findAll("span")
print(post_titless[0].text)
print(post_titless[2].text)
```

# Info

```python
intro_box = soup.find_all("div",{"class": "intro-box"})

poster = intro_box[0].div.a.img.get('src')
print(poster)

stars_rating = intro_box[0].div.findAll("a")
stars_rating2 = stars_rating[1].div.text
print(stars_rating2)

list_info = intro_box[0].div.ul.findAll("li")
country = list_info[1].text
video_duration = list_info[2].text
print(country)
print(video_duration)


list_info_2 = intro_box[0].div.findAll("ul")
list_info_22 = list_info_2[1].findAll("li")
list_info_23 = list_info_22[0].findAll("a")
release_date = list_info_23[0].text
release_year = list_info_23[1].text
category = list_info_22[3].a.text
print(release_date)
print(release_year)
print(category)


list_info_3 = intro_box[0].div.findAll("div",{"class": "columns"})
stars_rating2 = list_info_3[1].p
stars_rating3 = stars_rating2.text
print(stars_rating3)



```
