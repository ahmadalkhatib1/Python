```python
# conding: utf-8
from typing import Container
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'
headers = {'User-Agent':user_agent,} 

# https://lodynet.cc/%d9%85%d8%b3%d9%84%d8%b3%d9%84-%d9%81%d9%87%d9%85%d8%aa-%d8%b9%d9%8a%d9%88%d9%86%d9%83-%d8%ad%d9%84%d9%82%d8%a9-177-%d9%85%d8%aa%d8%b1%d8%ac%d9%85%d8%a9/
url = input('Enter Url :')

request = urllib.request.Request(url,None,headers)
response = urllib.request.urlopen(request)

html = response.read()
soup = bs(html, "html.parser")
# print(soup)
```

# Post Title

```python
post_title = soup.find_all("div",{"class": "TitleSingle"})
print(post_title[0].h1.text)
```
