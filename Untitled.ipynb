```python
from typing import Container
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'
headers = {'User-Agent':user_agent,} 

# url = ""
url = "https://lodynet.cc/%D9%85%D8%B3%D9%84%D8%B3%D9%84-%D8%A5%D9%8A%D9%87-%D8%AC%D8%A7%D8%A8%D9%83-%D8%B9%D9%86%D8%AF-%D8%AC%D8%A7%D8%B1%D9%83-%D8%A7%D9%84%D9%85%D9%88%D8%B3%D9%85-%D8%A7%D9%84%D8%AB%D8%A7%D9%86%D9%8A-13/"

request = urllib.request.Request(url,None,headers)
response = urllib.request.urlopen(request)

html = response.read()
soup = bs(html, "html.parser")
# print(soup)
```

# Post Title

```python
post_title = soup.find_all("div",{"class": "TitleSingle"})
print(post_title[0].h1.text)
```

# Post Video

```python
video_link = soup.find_all("ul",{"class": "ServersList"})
for video_link in video_link:
    for links in video_link.findAll("li"):
        print(links.text)
        print(links.get('data-embed'))
        print()
```
