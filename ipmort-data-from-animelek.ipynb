```python
from typing import Container
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.82 Safari/537.36'
headers = {'User-Agent':user_agent,} 

# url = ""
url = "https://animelek.me/episode/boku-no-hero-academia-5th-season-%D9%88%D8%A7%D9%84%D8%A7%D8%AE%D9%8A%D8%B1%D8%A925-%D8%A7%D9%84%D8%AD%D9%84%D9%82%D8%A9/"

request = urllib.request.Request(url,None,headers)
response = urllib.request.urlopen(request)

html = response.read()
soup = bs(html, "html.parser")
# print(soup)
```

# Post Title

```python
post_title = soup.find_all("h1",{"class": "anime-details-title"})
print(post_title[0].text)
```

# Post Videos

```python
video_link = soup.find_all("div",{"class": "hardsub-content"})
for video_link in video_link:
    for links in video_link.findAll("ul"):
        for links in video_link.findAll("li"):
            print(links.a.text)
            print(links.a.get('data-ep-url'))
            print()
```

# Post Downloads

```python
download_link = soup.find_all("div",{"id": "downloads"})
for download_link in download_link:
    for links in download_link.findAll("ul"):
        for links in download_link.findAll("li",{"class": "watch"}):
            print(links.a.text)
            print(links.a.get('href'))
            print()
```
